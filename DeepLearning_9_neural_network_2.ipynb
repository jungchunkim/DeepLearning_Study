## vanishing gradient



시그모이드에 의해 back propagation이 깊어질수록(layer가 많아질수록) 수가 계속 작아져서 gradient가 소실된 것과 같은 결과를 나타낸다.

#### 시그모이드가 1보다 작기 때문에 layer가 많아질수록 결국은 가장 작은 값을 나타낸다는것!! -> sigmoid 대신 ReLU를 사용해보자

#### ReLU를 사용하면 accuracy도 layer가 많아질수록 상승하고 cost도 학습할때마다 급격하게 떨어져 --> 좋음!!



# overfitting




training data에서는 높은 정확률을 보이지만 test data가 나왔을 때 저조한 정확률을 보임

## overfitting을 줄이는 방법

 - More training data
 - Reduce the number of features
 - Regularization
 - Dropout
 - Ensemble
 
### Regularization

weight를 큰 값읋 가지지 않게 하자 (꼬불꼬불하지 않게!)




### Dropout




학습할때는 랜덤하게 몇 개를 빼고 학습시킨다. -> test 할 때 전체 데이터를 실험해본다 --> 효과가 더 좋아

### ensemble



독립적으로 neural network를 만든 후에 나중에 합쳐서 ensemble prediction을 만들어주자
